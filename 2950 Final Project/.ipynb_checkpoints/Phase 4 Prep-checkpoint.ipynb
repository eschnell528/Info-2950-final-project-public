```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn import metrics
from sklearn.metrics import PrecisionRecallDisplay

from scipy import stats
```

```python
#import the dataframe
df = pd.read_csv('2950 Data Annotated.txt', sep='^', header = None)
df.columns = ['Link', 'Title', 'Author', 'Year', 'Authors Gender', 'Anger', 'Anticipation', 'Disgust', 'Fear', 'Joy', 'Negative', 'Positive', 'Sadness', 'Surprise', 'Trust', 'Number of Hits']
df=df.drop(['Link','Author'], axis=1)
#generate male and female dataframes
male_df = df[df['Authors Gender']=='m'].drop(['Authors Gender'], axis=1)
female_df = df[df['Authors Gender']=='f'].drop(['Authors Gender'], axis=1)
both_df = df.drop(['Authors Gender'], axis=1)
male_data = male_df.drop(['Title'], axis=1)
female_data = female_df.drop(['Title'], axis=1)
both_data = both_df.drop(['Title'], axis=1)
```

```python
#
for m in male_data.columns:
    if (m!='Year'):
        male_data[m] = male_data[m]/male_data['Number of Hits']
for f in female_data.columns:
    if (f!='Year'):
        female_data[f] = female_data[f]/female_data['Number of Hits']
for b in both_data.columns:
    if (b!='Year'):
        both_data[b] = both_data[b]/both_data['Number of Hits']
```

```python
model = LinearRegression().fit(trimmed_male_data['Year'].values.reshape(-1,1), trimmed_male_data['Sadness'].values.reshape(-1,1))
coef = model.coef_
print(coef)
print(model.intercept_)
```
